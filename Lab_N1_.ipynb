### Se importa las librerias, usaremos Pandas para cargar el dataframe

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Se crea el dataframe

```python
data = {
    'year': [2010, 2011, 2012, 2010, 2011, 2012, 2010, 2011, 2012],
    'team': [
        'FCBarcelona', 'FCBarcelona', 'FCBarcelona',
        'RMadrid', 'RMadrid', 'RMadrid',
        'ValenciaCF', 'ValenciaCF', 'ValenciaCF'
    ],
    'wins': [30, 28, 32, 29, 32, 26, 21, 17, 19],
    'draws': [6, 7, 4, 5, 4, 7, 8, 10, 8],
    'losses': [2, 3, 2, 4, 2, 5, 9, 11, 11]
}
```

### Se crea el DF football con las columnas year, team, wins, draws, losses. finalmente imprimimos el DF para visualizarlo

```python
football = pd.DataFrame(data, columns=['year', 'team', 'wins', 'draws', 'losses'])
print(football)
```

### Cargamos el DF leyendo el csv con pandas



```python
edu = pd.read_csv('/content/educ_figdp_1_Data.csv', na_values=':', usecols=["TIME", "GEO", "Value"])
print(edu.head())
print(edu.describe())
```

### Tomamos los valores de la posicion 10 a la 14

```python
slice_rows = edu[10:14]
print(slice_rows)
```

```python
filtered_data = edu[edu['Value'] > 6.5]
print(filtered_data.tail())
```

### Mostramos el grafico

```python
plt.figure(figsize=(10, 5))
plt.plot(football['year'], football['wins'], label='Wins')
plt.plot(football['year'], football['draws'], label='Draws')
plt.plot(football['year'], football['losses'], label='Losses')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.title('Football Team Performance')
plt.legend()
plt.show()
```
